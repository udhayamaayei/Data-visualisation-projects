```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Preprocessed_data.csv')
```

```python
# Histograms
df.hist(figsize=(20, 40), bins=30, edgecolor='black')
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(20, 10))
sns.boxplot(data=df.iloc[:,:-15])
plt.xticks(rotation=90)
plt.title('Box Plots of Protein Expression Levels')
plt.show()
```

```python
# Pie Chart
class_columns = ['class_c-CS-m', 'class_c-CS-s', 'class_c-SC-m', 'class_c-SC-s',
                 'class_t-CS-m', 'class_t-CS-s', 'class_t-SC-m', 'class_t-SC-s']

class_counts = df[class_columns].sum()

class_labels = ['Control, Stimulated, Memantine', 'Control, Stimulated, Saline',
                'Control, Not Stimulated, Memantine', 'Control, Not Stimulated, Saline',
                'Trisomic, Stimulated, Memantine', 'Trisomic, Stimulated, Saline',
                'Trisomic, Not Stimulated, Memantine', 'Trisomic, Not Stimulated, Saline']


plt.figure(figsize=(10, 7))
plt.pie(class_counts, labels=class_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Mice Classes')
plt.axis('equal')  
plt.show()
```

```python
# Count plot for Genotype
sns.countplot(data=df, x='Genotype_Control',color='green')
plt.title('Count Plot of Genotype')
plt.show()

# Count plot for Treatment
sns.countplot(data=df, x='Treatment_Saline',color='green')
plt.title('Count Plot of Treatment')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Desktop\\Preprocessed_data.csv")

protein_columns = df.columns[:77]  

class_columns = ['class_c-CS-m', 'class_c-CS-s', 'class_c-SC-m', 'class_c-SC-s',
                 'class_t-CS-m', 'class_t-CS-s', 'class_t-SC-m', 'class_t-SC-s']

mean_protein_by_class = pd.DataFrame(index=protein_columns, columns=class_columns)

for class_col in class_columns:
    class_df = df[df[class_col] == 1]
    mean_protein = class_df[protein_columns].mean()
    mean_protein_by_class[class_col] = mean_protein

mean_protein_by_class = mean_protein_by_class.T

mean_protein_by_class.fillna(0, inplace=True)

plt.figure(figsize=(20, 14))  
sns.heatmap(mean_protein_by_class.T, cmap='viridis', annot=False, linewidths=0.5)
plt.title('Heatmap of Protein Expression Levels by Class')
plt.xlabel('Classes')
plt.ylabel('Proteins')
plt.xticks(ticks=range(len(class_columns)), labels=class_columns, rotation=45, ha='right')  
plt.yticks(ticks=range(len(protein_columns)), labels=protein_columns) 
plt.show()
```

```python
# Heatmap
protein_columns = df.columns[0:22]

# Calculating the correlation matrix
correlation_matrix = df[protein_columns].corr()

# Plotting the heatmap
plt.figure(figsize=(16, 14))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Protein Expressions', fontsize=16)
plt.show()
```

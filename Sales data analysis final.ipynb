```python
df['Profit'] = df['Net Amount'] - df['Purchase Price']

# Step 2: Calculate Profit Margin
df['Profit Margin'] = (df['Profit'] / df['Net Amount']) * 100
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe is named df
df['Bill Date'] = pd.to_datetime(df['Bill Date'])  # Ensure 'Bill Date' is in datetime format

# Create a new column 'Month-Year' for grouping by month
df['Month-Year'] = df['Bill Date'].dt.to_period('M')  # This will give 'YYYY-MM' format

# Group by 'Month-Year' and sum the 'Profit' for each month
monthly_profit = df.groupby('Month-Year')['Profit'].sum()

# Plot the bar chart
plt.figure(figsize=(10, 6))
monthly_profit.plot(kind='bar', color='green')
plt.title('Monthly Profit')
plt.xlabel('Month-Year')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Ensure labels don't get cut off
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe is named df
df['Bill Date'] = pd.to_datetime(df['Bill Date'])  # Ensure 'Bill Date' is in datetime format

# Group by 'Bill Date' and sum the 'Profit' for each day
daily_profit = df.groupby('Bill Date')['Profit'].sum()

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(daily_profit, bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Daily Profits')
plt.xlabel('Profit')
plt.ylabel('Frequency (Number of Days)')
plt.grid(True)
plt.tight_layout()  # Ensure layout doesn't get cut off
plt.show()
```

```python
# Sales Trend Line Plot
df['Bill Date'] = pd.to_datetime(df['Bill Date'])
df.set_index('Bill Date', inplace=True)
monthly_sales = df['Net Amount'].resample('M').sum()

plt.figure(figsize=(10,6))
plt.plot(monthly_sales, label='Monthly Sales')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar plot for total sales by product
product_sales = df.groupby('Product Name')['Net Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
product_sales.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Products by Sales')
plt.ylabel('Sales Amount')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot for sales vs profit
plt.figure(figsize=(8,6))
sns.scatterplot(x='Net Amount', y='Profit', data=df)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

```python
# Heatmap of correlations between variables
plt.figure(figsize=(10,6))
corr = df[['Net Amount', 'Profit', 'Discount Amount', 'Quantity']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pie chart of sales by top customers
customer_sales = df.groupby('Customer Name')['Net Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,8))
plt.pie(customer_sales.head(10), labels=customer_sales.head(10).index, autopct='%1.1f%%', startangle=140)

plt.title('Top 10 Customers by Sales')
plt.show()
```

```python
# Pair plot of numerical columns
sns.pairplot(df[['Net Amount', 'Profit', 'Discount Amount', 'Quantity']])
plt.show()
```

```python
# Area plot for cumulative sales over time
plt.figure(figsize=(10,6))
monthly_sales.cumsum().plot(kind='area', color='lightblue', alpha=0.5)
plt.title('Cumulative Monthly Sales')
plt.ylabel('Cumulative Sales')
plt.show()
```

```python
# Count plot for the number of transactions by product
plt.figure(figsize=(12,6))
sns.countplot(x='Product Name', data=df, order=df['Product Name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Transaction Count per Product')
plt.show()
```

```python
from pandas.plotting import lag_plot

# Lag plot for sales to find patterns over time
plt.figure(figsize=(8,6))
lag_plot(monthly_sales)
plt.title('Lag Plot of Sales')
plt.show()
```

```python
product_sales = df['Product Name'].value_counts()

plt.figure(figsize=(10,6))
product_sales.plot(kind='bar')
plt.title('Sales Count by Product')
plt.xlabel('Product Name')
plt.ylabel('Count')
plt.show()
```

```python
profit = df[df["Profit"] > 0]

plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.histplot(profit['Profit'], bins=30, color='green', kde=True)
plt.title('Distribution of Profit-Making Sales Profits')
plt.xlabel('Profit')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
sns.histplot(loss['Profit'], bins=30, color='yellow', kde=True)
plt.title('Distribution of Loss-Making Sales Profits')
plt.xlabel('Loss')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
loss_by_product = loss['Product Name'].value_counts().head(20)

plt.figure(figsize=(12,6))

loss_by_product.plot(kind='bar', color='orange')
plt.title('Top 10 Products with Most Loss-Making Sales')
plt.xlabel('Product Name')
plt.ylabel('Number of Loss-Making Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
profit_by_product = profit['Product Name'].value_counts().head(20)

plt.figure(figsize=(12,6))

profit_by_product.plot(kind='bar', color='orange')
plt.title('Top 20 Products with Most Profit-Making Sales')
plt.xlabel('Product Name')
plt.ylabel('Number of Profit-Making Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation between discounts and profit
plt.figure(figsize=(8,6))
sns.scatterplot(x='Discount Amount', y='Profit', data=loss, alpha=0.6)
plt.title('Profit vs Discount Amount')
plt.xlabel('Discount Amount')
plt.ylabel('Profit')
plt.show()

# Calculate correlation coefficient
correlation = loss['Discount Amount'].corr(loss['Profit'])
print(f"Correlation between Discount Amount and Profit: {correlation:.2f}")
```
